### Basis Vector & Linear Transformations

  *   Basis Vector: A set of vectors that define a coordinate system. Any vector in space can be represented as a combination of basis vectors.

  *   Linear Transformations: A function that maps vectors to other vectors while preserving vector addition and scalar multiplication.

 *    Example: Scaling, Rotation, and Reflection are linear transformations.

###  Linear Transformation in 3D & Matrix Multiplication

  *   Linear Transformation in 3D: Extends 2D transformations to 3D space using 3x3 matrices.

  *   Matrix Multiplication as Composition: Combining multiple transformations by multiplying their matrices.

 *    Test of Commutative Law:

  *       Matrix multiplication is NOT commutative → AB ≠ BA in general.

     *    Example: Rotation followed by scaling ≠ Scaling followed by rotation.

## Determinants & Inverse Matrices

  *   Determinant: A scalar value that provides information about a matrix’s properties (e.g., area scaling factor, invertibility).

  *   Inverse of a Matrix: Exists only for square matrices with a non-zero determinant (non-singular matrices).

 *    Why only square matrices have an inverse?

   *      Because inverse operation requires the matrix to have the same number of rows and columns.

  *   Why inverse is possible only for non-singular matrices?

      *  A singular matrix (det = 0) collapses dimensions, making it non-reversible.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a transformation matrix (Scaling & Rotation)
transformation_matrix = np.array([[1.5, -0.5], [0.5, 1.5]])

# Define basis vectors
basis_vectors = np.array([[1, 0], [0, 1]])

# Apply transformation
transformed_vectors = np.dot(transformation_matrix, basis_vectors)

# Plot the vectors
plt.quiver([0, 0], [0, 0], transformed_vectors[0], transformed_vectors[1], color=['r', 'b'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.title("Linear Transformation")
plt.show()
```

### Introduction to Machine Learning
4️⃣ Types of Machine Learning

*   Supervised Learning: Uses labeled data for training.

      *   Examples: Regression (Predicting house prices), Classification (Spam detection).

* Unsupervised Learning: Works with unlabeled data to find patterns.

    *  Examples: Clustering (Customer segmentation), Anomaly Detection (Fraud detection).

* Reinforcement Learning: Learns through trial and error, receiving rewards or penalties.
  *  Examples: Game AI, Robotics.

6️⃣ Data Mismatch

    Issue: The distribution of training data is different from real-world test data.

    Solution: Collect diverse data, use domain adaptation techniques.

Challenges in Machine Learning

1️⃣ Poor Quality of Data

    Issue: Noisy, missing, or incorrect data can mislead models.

    Solution: Data cleaning, handling missing values, outlier detection.

2️⃣ Irrelevant Features

    Issue: Unnecessary data makes learning harder and increases computational cost.

    Solution: Feature selection, dimensionality reduction.

3️⃣ Insufficient Quantity of Training Data

    Issue: Too little data makes it hard for models to generalize well.

    Solution: Data augmentation, synthetic data generation, transfer learning.

4️⃣ Non-Representative Training Data

    Issue: If the dataset doesn't reflect real-world scenarios, the model will be biased.

    Solution: Ensure diverse, balanced, and representative samples.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the linear transformation matrix (e.g., scale by 2)
transformation_matrix = np.array([[2, 0], [0, 2]])  # Scaling by 2 in both x and y directions

# Original vector
v = np.array([3, 4])  # Example vector (3, 4)

# Apply the linear transformation
transformed_v = np.dot(transformation_matrix, v)

# Plotting the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="Original Vector")
plt.quiver(0, 0, transformed_v[0], transformed_v[1], angles='xy', scale_units='xy', scale=1, color='b', label="Transformed Vector")

# Labels and grid
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
