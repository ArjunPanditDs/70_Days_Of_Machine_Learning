### Simple Linear Regression First Model

```python
# importing the required library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# loading the dataset into dataframe
df=pd.read_csv("/content/synthetic_linear_regression_data.csv")
```

```python
# Getting the first five row of the data
df.head()
```

```python
# Getting the overview of the data
df.info()
```

```python
# Getting the shape of the data
df.shape
```

```python
# plotting the shape of the data
plt.scatter(df['Hours'],df['Scores'])
plt.xlabel("Hours of study")
plt.ylabel("Scores of Students")
plt.plot()
```

```python
# Separating the input and output column for model
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
# Split data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
# Making the object of LinearRegression
lr=LinearRegression()
```

```python
# Traing the Linear Model
lr.fit(x_train,y_train)
```

```python
# Our test data of input
x_test
```

```python
# our test data of output
y_test
```

```python
# Predicting with input value
lr.predict(x_test.iloc[[0]])
```

```python
# Predicting with input value
lr.predict(x_test.iloc[[1]])
```

```python
# Finding the straight line
plt.scatter(df['Hours'],df['Scores'])
plt.plot(x_train,lr.predict(x_train),color='yellow')
plt.xlabel("Hours of study")
plt.ylabel("Scores of Students")
```

```python
# Getting the Tan theta
m=lr.coef_
```

```python
# intercet on y
c=lr.intercept_
```

```python
# doing predcition on basis of formula
# y=mx+c

y_pred=m*3.44+c
y_pred
```
